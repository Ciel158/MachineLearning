```python
import numpy as np
from math import log

x = np.arange(0.1, 5., 0.1) # range()와는 달리 실제로 값이 생성되어 ndarray로 저장
plt.plot(x, np.log(x), 'r') #빨간색 그리기. 
plt.plot(x, x*0, 'b--') #축 , 파란색+점선 그리기
plt.plot(x, np.exp(x), 'b') #자수함수 파란색으로 그리기
plt.legend(['log', 'exp']) # 그래프를 그린 순서에 따라 범례를 표시
plt.show()

import numpy as np
from math import log #값이 실수이거나 정수여야지만 돌아감 

x = np.arange(0.1, 2., 0.1) # range()와는 달리 실제로 값이 생성되어 ndarray로 저장
plt.plot(x, np.log(x), 'r', label="log")
plt.plot(x, x*0, 'b--', label='0')
plt.plot(x, np.exp(x), 'g',label='exp')
plt.legend(['log', '0', 'exp']) # 그래프를 그린 순서에 따라 범례를 표시
plt.show()
```

```python
# 수평 막대 그래프
names = ['a', 'b', 'c', 'd']
weights = [25, 46, 28, 34]

plt.barh(names, weights) 
plt.show()

# 기본적으로 각 막대는 원점에서부터 순서대로 그려진다. 
# 따라서 현재 상태로 그리면 위에서부터 역순으로 그려진 것처럼 보인다.
#아래와 같이 names와 weights를 역순으로 바꿔서 그리면 위에서부터 순서대로 그릴 수 있다.
plt.barh(names[::-1], weights[::-1])
plt.show()

sorted_w = sorted(zip(names, weights), key=lambda kv: kv[1]) #zip>튜플, 묶는 이유>X축 Y축이 연결되어있지 않음 이름값쌍이 변화
n, w = zip(*sorted_w) #*>언패킹,원점에서부터 높은 게 되므로 높은 게 가장 위에 가고 싶으면 리버스가 없는 게 맞음 
#n, w = zip(sorted_w)
plt.barh(n, w, height=0.5, color='purple')
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs' #각 파이의 label 현재 튜플로 집어넣음 리스트도 가능 튜플>변경불가 리스트>변경가능
sizes = [15, 30, 45, 10] #각 파이의 크기
explode = (0, 0.1, 0, 0)  # 각 파이에 대해 중앙으로부터의 거리 지정, 여기서는 두번째 파이를 강조하기 위해 떼어냄 현재 튜플로 집어넣음 리스트도 가능 튜플>변경불가 리스트>변경가능

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, counterclock=False, startangle=90)
#autopct: 파이의 값을 보여주기 위한 포맷팅, 소숫점이하 한 자리 %(실제내용)%로 가둠
#shadow: 그림자 여부
#startangle: 시작 위치를 지정, 90이면 y축 양의 방향(12시)에서 출발 디폴트>3시방향 
#counterclock : True>반시계방향 False>시계방향

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
#정규분포 >모든 정규분포를 실험가능. 
#random 함수를 사용할 때, reproducibility(재생산성)을 위해 seed를 지정. seed 사용 이유>random을 할 때 남이 해봤던 값으로 시험해볼 수 없음==재생산성. 난수표    
np.random.seed(19680801) 

height, sigma = 170, 10
x = height + sigma * np.random.randn(10000) #randn>정규분포형식으로 뽑음, 10000개의 값을 생성, 표준편차>밀집도>3시그마를 잘 안넘음, 균일한 애들==유니폼 분포

# bins는 나누어서 보여 줄 칸의 수
# density가 True이면 y축을 확률로 변경, False이면 Count
plt.hist(x, bins=50, density=True, facecolor='g', alpha=0.5) # facecolor:내부색깔, alpha:투명도 조정
#plt.hist(x, bins=50)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('키 히스토그램')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') # (60, 0.025) 좌표에 텍스트를 출력, $로 감싸서 TeX 수식 사용 가능
#plt.axis([40, 160, 0, 0.03]) #y축을 확률로 가정하고 적절한 범위를 지정, 만일 count로 하게 되면 조정이 필요
plt.grid(True)
plt.show()
```
